```python
df = merged.merge(Final_demo, on='client_id', how='inner')
df.drop_duplicates()
display(df)
```

```python
patient = df[['client_id', 'clnt_age', 'gendr', 'Variation', 'age_group', 'process_step']]
patient.drop_duplicates()
```

```python
Test = patient[patient['Variation'] == 'Test']
Control = patient[patient['Variation'] == 'Control']
print(Test)
```

```python
print(Control)
```

```python
Test['age_group'], Test['gendr'].value_counts().sort_index()
```

```python
Control['age_group'], Control['gendr'].value_counts().sort_index()
```

```python
from scipy.stats import chi2_contingency
crosstab_result1 = pd.crosstab(Test['age_group'], Test['gendr'])

chi2_statistic, chi2_p_value, _, expected_freq = chi2_contingency(crosstab_result1)

chi2_statistic, chi2_p_value, expected_freq
from scipy.stats.contingency import association

association(crosstab_result1, method="cramer")
crosstab_result1.plot(kind="bar", stacked=True);
plt.show()
```

```python
from scipy.stats import chi2_contingency
crosstab_result2 = pd.crosstab(Control['age_group'], Control['gendr'])

chi2_statistic, chi2_p_value, _, expected_freq = chi2_contingency(crosstab_result2)

chi2_statistic, chi2_p_value, expected_freq
from scipy.stats.contingency import association

association(crosstab_result2, method="cramer")
crosstab_result2.plot(kind="bar", stacked=True);
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(Test['clnt_age'], label='Test')
sns.kdeplot(Control['clnt_age'], label='Control')
plt.legend()
plt.title('Age Distribution Comparison')
plt.show()
```

```python
age_counts = pd.DataFrame({
    'Test': Test['age_group'].value_counts(),
    'Control': Control['age_group'].value_counts()
}).sort_index()
age_counts.plot(kind='bar', figsize=(10, 6), title='Age Group Comparison')
plt.xlabel('Age group')
plt.ylabel('No. clients')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
