```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
Web_data_1 = pd.read_csv('../data/raw/df_final_web_data_pt_1.txt')
Web_data_2 = pd.read_csv('../data/raw/df_final_web_data_pt_2.txt')
Final_demo = pd.read_csv('../data/raw/df_final_demo.txt')
Clients = pd.read_csv('../data/raw/df_final_experiment_clients.txt')
display(Web_data_1)
```

```python
combined_df = pd.concat([Web_data_1, Web_data_2], ignore_index=True)
combined_df.drop_duplicates()
```

```python
display(combined_df)
```

```python
merged = combined_df.merge(Clients, on='client_id', how='inner')
```

```python
display(merged)
```

```python
df = merged.merge(Final_demo, on='client_id', how='inner')
df.drop_duplicates()
display(df)
```

```python
grouped_clients = df.groupby("client_id").agg({"bal" : ["sum", "mean"], "calls_6_mnth" : ["sum","mean"]})
grouped_clients
```

```python
df.columns
```

```python
grouped_clients.columns = ['bal_sum', 'bal_mean', 'calls_sum', 'calls_mean']

# Sort by a specific column — for example, total balance
grouped_clients_sorted = grouped_clients.sort_values(by='bal_sum', ascending=False)

# Plot
grouped_clients_sorted['bal_sum'].plot(kind='bar', figsize=(10, 5))
plt.title("Total Balance per Client")
plt.ylabel("Balance")
plt.xlabel("Client ID")
plt.tight_layout()
plt.show()
```

```python
grouped_clients[['bal_mean', 'calls_mean']].plot(kind='line', figsize=(10, 5))
plt.title("Mean Balance and Mean Calls per Client")
plt.xlabel("Client ID")
plt.ylabel("Mean Value")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=grouped_clients, x='calls_sum', y='bal_sum')
plt.title("Total Calls vs Total Balance")
plt.xlabel("Total Calls (6 months)")
plt.ylabel("Total Balance")
plt.tight_layout()
plt.show()
```

```python
grouped_clients['bal_mean'].plot(kind='hist', bins=20, edgecolor='black')
plt.title("Distribution of Mean Balance")
plt.xlabel("Mean Balance")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=grouped_clients[['bal_mean', 'calls_mean']])
plt.title("Boxplot of Mean Balance and Mean Calls")
plt.tight_layout()
plt.show()
```

```python
df_test_unique = (
    df[df['Variation'] == 'Test'].sort_values(by='date_time').groupby('client_id').last().reset_index())

df_control_unique = (df[df['Variation'] == 'Control'].sort_values(by='date_time').groupby('client_id').last().reset_index())
```

```python
df_test_unique['logons_6_mnth'].describe()
df_control_unique['logons_6_mnth'].describe()
df_test_unique['calls_6_mnth'].describe()
df_control_unique['calls_6_mnth'].describe()
```

```python
def engagement_level(row):
    score = (
        row['logons_6_mnth'] +
        row['calls_6_mnth'] +
        row['num_accts']
    )
    if score >= 20:
        return 'High'
    elif score >= 10:
        return 'Medium'
    else:
        return 'Low'

df['engagement_level'] = df.apply(engagement_level, axis=1)
```

```python
df_control_unique['engagement_level'] = df_control_unique.apply(engagement_level, axis=1)
df_test_unique['engagement_level'] = df_control_unique.apply(engagement_level, axis=1)
```

```python
test_counts = df_test_unique['engagement_level'].value_counts()
control_counts = df_control_unique['engagement_level'].value_counts()

engagement_compare = pd.DataFrame({
    'Test Group (%)': (test_counts/23532).round(1),
    'Control Group (%)': (control_counts/23535).round(1)
})

print(engagement_compare)
```
