```python
clients_with_error = df_control[df_control['visit_id'].isin(visits_without_confirm)]['client_id'].unique()
```

Unique clients in control

```python
total_clients = df_control['client_id'].nunique()
total_clients
```

```python
qty_client_error = len(clients_with_error)
```

```python
percentage_error = (qty_client_error/total_clients) * 100
percentage_error
```
